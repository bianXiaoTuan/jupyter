# Grouping by another series

```python
life = pd.read_csv('data/life_expectancy.csv', index_col='Country')
life.head()
```

```python
regions = pd.read_csv('data/regions.csv', index_col='Country')
regions.head()
```

```python
life.groupby(regions['region']).mean()
```

```python
life.groupby(regions['region'])['2010'].mean()
```

# Computing multiple aggregates of multiple columns

```python
df = pd.read_csv('data/titanic.csv')
df.head()
```

```python
df.groupby('Pclass')[['Age', 'Fare']].agg(['max', 'median'])
```

```python
df.groupby('Pclass')[['Age', 'Fare']].agg(['max', 'median']).loc[:, ('Age', 'max')]
```

```python
df.groupby('Pclass')[['Age', 'Fare']].agg(['max', 'median']).loc[:, ('Fare', 'median')]
```

# Aggregating on index levels/fields

```python
df = pd.read_csv('data/gapminder.csv', index_col=['Year', 'region', 'Country']).sort_index()
df.head()
```

```python
by_year_region = df.groupby(level=['Year', 'region'])
by_year_region
```

```python
# Define the function to compute spread: spread
def spread(series):
    return series.max() - series.min()

# Create the dictionary: aggregator
aggregator = {'population':'sum', 'child_mortality':'mean', 'gdp':spread}
```

```python
by_year_region.agg(aggregator)
```

# Grouping on a function of index

```python
df = pd.read_csv('data/new_sales.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
time = df.index.strftime('%a')
time
```

```python
df.groupby(time)['Units'].sum()
```
